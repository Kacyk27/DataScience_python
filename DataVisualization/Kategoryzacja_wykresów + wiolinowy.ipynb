```python
import plotly.express as px
from seaborn import load_dataset
import pandas as pd
import plotly.figure_factory as ff
import plotly.graph_objects as go
```

```python
df = load_dataset('titanic')
df.head()
```

```python
px.scatter(df,x='age',y='fare',color='survived', facet_col='sex', facet_row='class',
           category_orders={'class':['First','Second','Third'], 'sex':['female','male']})
```

```python
px.violin(df,x='survived',y='age',color='survived', points='all', box=True)
```

```python
px.violin(df,y='age',color='survived', violinmode='overlay')
```
