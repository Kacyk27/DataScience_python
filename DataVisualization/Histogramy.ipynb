```python
import plotly.express as px
from seaborn import load_dataset
import pandas as pd
import plotly.figure_factory as ff
import plotly.graph_objects as go
```

```python
df = load_dataset('mpg')
df.head()
```

```python
df.describe()
```

```python
df.origin.value_counts()
```

```python
df.info()
```

```python
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmiennej mpg')
```

```python
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmiennej mpg', histnorm='probability density')
```

```python
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmiennej mpg', histnorm='probability density',
             facet_row='origin', color='origin')
```

```python
dff = df[(df.origin == 'usa') | (df.origin == 'europe')]
px.histogram(dff, x='mpg', nbins=50, title='Rozkład zmiennej mpg', color='origin',
             marginal='box', opacity=0.7)
```

```python
px.histogram(df, x='origin')
```

```python
fig = ff.create_distplot([df.mpg], group_labels=['mpg'])
fig.show()
```

```python
fig = ff.create_distplot([df.mpg, df.acceleration], group_labels=['mpg', 'Acceleration'])
fig.show()
```

#Histogram 2D

```python
df = load_dataset('mpg')
df.head()
```

```python
fig = go.Figure(go.Histogram2d(
    x=df.mpg,
    y=df.acceleration,
    nbinsx=30,
    nbinsy=30,
))

fig.add_trace(go.Scatter(
    x=df.mpg,
    y=df.acceleration,
    mode='markers',
    marker={
        'symbol':'x',
        'opacity':0.7,
        'color':'white',
        'size':8,
        'line':{
            'width': 1
        }
    }
))

fig.show()
```
