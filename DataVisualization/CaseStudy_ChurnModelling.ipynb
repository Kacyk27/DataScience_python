#Churn Modelling

#### churn modelling - przewidywanie odejścia klienta aby temu zapobiec. Jest to tańsze, niż pozyskanie nowego

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


url = 'https://ml-repository-krakers.s3-eu-west-1.amazonaws.com/kaggle+/churn_modelling/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

#Wstępne przetworzenie

```python
TotalChargesMedian = df[df.TotalCharges != ' ']['TotalCharges'].astype('float32').median()
```

```python
df['TotalCharges'][df['TotalCharges'] == ' '] = TotalChargesMedian
```

```python
df.TotalCharges.value_counts()
```

```python
df.info()
```

```python
df['TotalCharges'] = df['TotalCharges'].astype('float32')
```

```python
df.info()
```

```python
if not df.isnull().sum().sum():
    print('Brak pustych danych')
```

```python
categorical = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 
               'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
               'StreamingTV', 'Contract', 'StreamingMovies', 'PaperlessBilling', 'PaymentMethod', 'Churn']
numerical = ['tenure', 'MonthlyCharges', 'TotalCharges']

for col in categorical:
    df[col] = pd.Categorical(df[col])

for col in numerical:
    df[col] = df[col].astype('float32')
    
```

```python
df.info()
```

```python
df = df.drop(columns=['customerID'])
df.info()
```

```python
df.describe()
```

```python
df.describe(include=['category'])
```

# Wykres rozkładu Churn

```python
plt.figure(figsize=(8,6), )
sns.set()
df.Churn.value_counts().plot(kind='pie', fontsize=16, colors=['#3DDC97', '#FF495C'], 
                             explode = [0.1,0], shadow=True, autopct='%1.1f%%')

plt.legend()
plt.ylabel=("")
plt.title('Rozkład Churn')
print(df.Churn.value_counts())
```

```python
plt.figure(figsize=(8,6), )

df.gender.value_counts().plot(kind='pie', fontsize=16, colors=['#3DDC97', '#FF495C'], 
                             explode = [0.1,0], shadow=True, autopct='%1.1f%%', labels=['',''])

plt.legend(['Male','Female'])
plt.ylabel=("")
plt.title('Rozkład Płci')
print(df.gender.value_counts())
```

```python
plt.figure(figsize=(8,6), )

df.SeniorCitizen.value_counts().plot(kind='pie', fontsize=16, colors=['#3DDC97', '#FF495C'], 
                             explode = [0.1,0], shadow=True, autopct='%1.1f%%', labels=['',''])

plt.legend(['0','1'])
plt.ylabel=("")
plt.title('Rozkład zmiennej SeniorCitizen')
print(df.SeniorCitizen.value_counts())
```
