```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
fig = make_subplots(rows=1, cols=2, shared_yaxes=True, shared_xaxes=True, 
                    subplot_titles=['first','second'], column_widths=[0.7,0.3])

fig.update_xaxes(title_text='x value', row=1, col=1)
fig.update_xaxes(title_text='x value', row=1, col=2)
fig.update_yaxes(title_text='y value', row=1, col=1)
fig.update_yaxes(title_text='y value', row=1, col=2)



fig.add_trace(go.Scatter(x=[1,2,3],
                         y=[2,3,1],),
              row=1, col=1)
fig.add_trace(go.Scatter(x=[1,2,3],
                         y=[8,1,2],),
              row=1, col=2)

fig.update_layout(title_text='Subplots')
fig.show()
```

```python
fig = make_subplots(rows=1, cols=2, shared_yaxes=True, shared_xaxes=True, 
                    subplot_titles=['first','second'], column_widths=[0.7,0.3],
                    specs=[[{'type':'xy'},{"type":'domain'}]])

fig.update_xaxes(title_text='x value', row=1, col=1)
fig.update_xaxes(title_text='x value', row=1, col=2)
fig.update_yaxes(title_text='y value', row=1, col=1)
fig.update_yaxes(title_text='y value', row=1, col=2)



fig.add_trace(go.Scatter(x=[1,2,3],
                         y=[2,3,1],),
              row=1, col=1)
fig.add_trace(go.Pie(values=[40,30,30]),
              row=1, col=2)

fig.update_layout(title_text='Subplots')
fig.show()
```
