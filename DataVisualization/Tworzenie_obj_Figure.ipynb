```python
trace0 = go.Bar(y=[2,3,1])

data = [trace0]

layout = go.Layout(title_text='Wykres słupkowy')

fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
trace0 = go.Bar(y=[2,3,1])
trace1 = go.Bar(y=[4,1,6])

data = [trace0,trace1]

layout = go.Layout(title_text='Wykres słupkowy')

fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
trace0 = go.Bar(y=[2,3,1])
trace1 = go.Bar(y=[4,1,6])
trace2 = go.Bar(y=[3,2,8])

data = [trace0,trace1,trace2]

layout = go.Layout(title_text='Wykres słupkowy')

fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
trace0 = go.Bar(x=[1,2,3], y=[2,3,1])
trace1 = go.Line(x=[1,2,3], y=[2,3,1])
data = [trace0, trace1]

layout = go.Layout(title_text='mieszany wykres')

fig = go.Figure(data, layout)
fig.show()
```

```python
trace0 = go.Bar(x=[1,2,3], y=[2,3,1])
trace1 = go.Line(x=[1,2,3], y=[2,3,1])
trace2 = go.Scatter(x=[1,2,3], y=[1,2,4], fill='tozeroy')

data = [trace0, trace1,trace2]

layout = go.Layout(title_text='mieszany wykres')

fig = go.Figure(data, layout)
fig.show()
```

```python
fig = go.Figure(
    data=[
          go.Bar(
              x=[1,2,3],
              y=[2,1,3]
          ),
          go.Line(
              x=[1,2,3],
              y=[4,5,4]
          ),
          go.Scatter(
              x=[1,2,3],
              y=[5,3,2],
              fill='tozeroy'
          )
    ],
    layout=go.Layout(
        title_text='mieszany, budowa od podstaw'
    )
)
fig.show()
```
