# Wykresy rozkładów - joinplot

```python
sns.set(style='darkgrid')

df = np.random.multivariate_normal(mean=[0,1], cov=[(1,-0.4), (-0.4, 1)], size=200)
df = pd.DataFrame(df, columns=['x','y'])
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.jointplot(data=df, x='x',y='y')
```

```python
sns.set(style='white')
sns.jointplot(data=df, x='x',y='y', kind='hex')
```

```python
sns.jointplot(data=df, x='x',y='y', kind='kde')
```

```python
g = sns.jointplot(data=df, x='x',y='y', kind='kde', n_levels=15, cmap='plasma')
g.plot_joint(plt.scatter, c='black', s=30, marker='+')
```

#Pairplot

```python
sns.set()
df = sns.load_dataset('iris')
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='species')
```

```python
sns.pairplot(df, hue='species', palette='viridis')
```

```python
g = sns.PairGrid(df, hue='species')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6)
```
