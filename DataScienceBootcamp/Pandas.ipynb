```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df = df[df.Quantity >=0]
df.info()
```

```python
df.describe()
```

```python
df.CustomerID = df.CustomerID.apply(lambda x: str(int(x)))
df.info()
```

### Dane przygotowane, czas porwna do SQL


wskazywanie na zasadzie wynikw: SQL == python

```python
# SELECT * FROM df;
df
```

```python
# SELECT Quantity, UnitPrice, CustomerID FROM df;
df[['Quantity','UnitPrice','CustomerID']]
```

```python
# SELECT Quantity, UnitPrice, CustomerID FROM df LIMIT 10;
df[['Quantity','UnitPrice','CustomerID']].head(10)
```

```python
# SELECT * FROM df WHERE CustomerID='17850';
df[df.CustomerID == '17850']
```

```python
# SELECT * FROM df WHERE CustomerID='17850' and UnitPrice > 5;
df[(df.CustomerID == '17850') & (df.UnitPrice>5)]
```

```python
# SELECT * FROM df WHERE CustomerID='17850' or Country='France';
df[(df.CustomerID == '17850') | (df.Country=='France')]
```

```python
# SELECT * FROM df WHERE InvoiceNo is not null;
df[df.InvoiceNo.notnull()]
```

```python
# SELECT CustomerID, count(*) FROM df GROUP BY CustomerID;
df.groupby('CustomerID').size()
```

```python
df['Revenue'] = df.Quantity * df.UnitPrice
df.head(2)
```

```python
# SELECT CustomerID, avg(Revenue), count(*) FROM df GROUP BY CustomerID;
df.groupby('CustomerID').aggregate({'Revenue':np.mean,'CustomerID':np.size})
```

```python
df.groupby('CustomerID').aggregate({'Revenue':np.mean,'CustomerID':np.size}).rename(
    columns={'Revenue':"RevenueAverage",'CustomerID':'CustomerID_count'}
)
```

```python
df['InvoiceDateDay'] = df.InvoiceDate.dt.day
df.head(2)
```

```python
#SELECT InvoiceDateDay, sum(Revenue) FROM df GROUP BY InvoiceDateDay
results = df.groupby('InvoiceDateDay').aggregate({'Revenue':np.sum}).rename(columns={'Revenue':'RevenueSum'})
results
```

```python
results.plot(kind='bar')
```
