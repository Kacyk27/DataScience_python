## Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

### Wykresy liniowe

```python
_ = plt.plot([0,1],[0,3])
```

```python
plt.plot([0,1],[0,3], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Tytuł wykresu')
_ = plt.legend()
```

```python
plt.plot([0,1,2],[0,3,2], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Tytuł wykresu')
_ = plt.legend()
```

```python
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

### Tworzenie wielu wykresów

```python
fig = plt.figure(figsize=(8,6))

plt.subplot(131)   #jedynka podaje liczbę wierszy w siatce, dwójka ile kolumn, jedynka do kolejnego wykresu
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([1,3,2],[2,3,5], 'r--')
plt.subplot(133)
plt.plot([0,1,2],[3,1,4], 'ro')
```

### wykresy słupkowe

```python
plt.bar(x=[0,1,2,3], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.bar(x=['a','b','c','d'], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.barh(y=['a','b','c','d'], width=[1.2, 3.2, 1.2, 5.2])
```

### wykresy punktowe

```python
from numpy.core.fromnumeric import size
x1 = np.random.randn(300)
x2 = np.random.randn(300)
x3 = np.random.randn(300) * 50

plt.scatter(x=x1,y=x2, s=x3)
```

```python
plt.scatter(x=x1,y=x2, s=x3, alpha=0.5)
```

```python
fig = plt.figure(figsize=(12,10))

plt.subplot(221)
plt.imshow(mpimg.imread('134383.jpg'))
plt.axis('Off')

plt.subplot(222)
plt.imshow(mpimg.imread('156487.jpg'))
plt.axis('Off')

plt.subplot(223)
plt.imshow(mpimg.imread('fallen-angel-wallpaper-for-desktop_2558512.jpg'))
plt.axis('Off')

plt.subplot(224)
plt.imshow(mpimg.imread('thumb-1920-480417.jpg'))
plt.axis('Off')
```

```python
fig = plt.figure(figsize=(12,10))

for idx,img in enumerate(['134383.jpg','156487.jpg','fallen-angel-wallpaper-for-desktop_2558512.jpg','thumb-1920-480417.jpg']):
    plt.subplot(220 + idx + 1)
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')

```
